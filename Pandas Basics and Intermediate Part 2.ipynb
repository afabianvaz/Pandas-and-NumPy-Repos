# Advanced Filtering with between(), isin() and ~

```python
summer = pd.read_csv('summer.csv')
summer.head()
```

```python
og_1988 = summer.loc[summer.Year == 1988]
og_1988.head()
```

```python
og_1988.tail()
```

```python
og_1988.info()
```

```python
og_1992 = summer.loc[summer.Year >= 1992]
og_1992.head()
```

```python
og_1992.tail()
```

```python
summer.Year.between(1996,1969).head()
```

```python
og_60s = summer.loc[summer.Year.between(1960,1969, inclusive = True)]
```

```python
og_60s.head()
```

```python
og_60s.tail()
```

```python
my_favouite_game= [1972, 1996]

summer.Year.isin(my_favouite_game).head()
```

```python
og_72_96 = summer.loc[summer.Year.isin(my_favouite_game)]
```

```python
og_72_96.head()
```

```python
og_72_96.tail()
```

```python
og_not_72_96 = summer.loc[~summer.Year.isin(my_favouite_game)]
og_not_72_96.head()
```

```python
og_not_72_96.tail()
```

```python
og_not_72_96.Year.unique()
```
