# Slicing Rows and Columns with Iloc

```python
summer.head(3)
```

```python
summer.iloc[0,4]
```

```python
summer.iloc[0, 0:5] # we can ectraxt five features from row 1
```

```python
summer.iloc[0, [2,3,4,7,8]] # we can ectraxt 2,3,4,7,8th features from row 1
```

```python
summer.iloc[34:39, [2,3,4,7,8]] # we can ectraxt 2,3,4,7,8th features from rows 34 to 38
```

```python
summer.iloc[:,4]
```

```python
summer.iloc[:,4].equals(summer.Athlete)
```

```python
summer['Athlete']
```
